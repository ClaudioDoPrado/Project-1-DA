```python
# Identify columns to drop (like 'Unnamed' patterns)
columns_to_drop = [col for col in df.columns if "Unnamed" in col]

# Drop the columns
df_cleaned = df.drop(columns=columns_to_drop)

# Display the cleaned DataFrame head
df_cleaned.head()
```

```python
# Dropping rows by index (0,1,10,30) also a range of rows from 32 to 38
df_dropped = df_cleaned.drop(index=[0,1,10,30]+list(range(32, 39)))
df_dropped
```

```python
# Deleting the Index
df_dropped.reset_index(drop=True, inplace=True)
df_dropped.head()
```

```python
# renaming a columns, Time -> Country
df_dropped.rename(columns={'TIME': 'Country'}, inplace=True)
df_dropped
```

```python
# Create DataFrame for the new types
df_dropped = pd.DataFrame(df_dropped)

# Select the columns to convert 
columns_to_convert = ['2010', '2011', '2012', '2013', '2014', 
                      '2015', '2016', '2017', '2018', '2019', 
                      '2020', '2021', '2022', '2023']

# Convert the selected columns to numeric, forcing invalid parsing to NaN
df_dropped[columns_to_convert] = df_dropped[columns_to_convert].apply(pd.to_numeric, errors='coerce')

df_dropped.dtypes
```

```python
# Rechecking how many duplicated rows is there.
df_dropped.isna().sum()
```

```python
# Create DataFrame
df_dropped = pd.DataFrame(df_dropped)

# Calculate year-on-year percentage change
profitability = df_dropped.iloc[:, 1:].pct_change(axis=1) * 100

# Separate pre-pandemic and post-pandemic data
pre_pandemic = profitability.iloc[:, 0:10]  # 2010-2019
post_pandemic = profitability.iloc[:, 10:]   # 2020-2023

# Compute average profitability before and after the pandemic
avg_pre_pandemic = pre_pandemic.mean(axis=1).rename("Avg Pre-Pandemic (%)")
avg_post_pandemic = post_pandemic.mean(axis=1).rename("Avg Post-Pandemic (%)")

# Combine results into a DataFrame for comparison
comparison = pd.concat([avg_pre_pandemic, avg_post_pandemic], axis=1)

# Display the comparison
print(comparison)

# Identify notable differences
comparison['Difference'] = comparison['Avg Post-Pandemic (%)'] - comparison['Avg Pre-Pandemic (%)']
print("\nNotable Differences:")
print(comparison[comparison['Difference'] > 0])  # Positive differences (increased profitability)
print(comparison[comparison['Difference'] < 0])  # Negative differences (decreased profitability)
```

```python
#DataFrame
df = pd.DataFrame(df_dropped)

# Plotting the data
plt.figure(figsize=(14, 7))
x = range(len(df["Country"]))
plt.bar(x, df["2019"], width=0.4, label="2019", color='b', align='center')
plt.bar([p + 0.4 for p in x], df["2022"], width=0.4, label="2022", color='g', align='center')

plt.xlabel('Country')
plt.ylabel('Prices Index')
plt.title('Prices Before and After Pandemic (2019 vs 2022)')
plt.xticks([p + 0.2 for p in x], df["Country"], rotation=90)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Ensure 'Country' is a string
df_dropped['Country'] = df_dropped['Country'].astype(str)
# Convert '2020' to numeric (if not already)
df_dropped['2020'] = pd.to_numeric(df_dropped['2020'], errors='coerce')
# Drop NaN values in '2020'
df_dropped = df_dropped.dropna(subset=['2020'])
# Plot using barplot instead of countplot
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='2020', data=df_dropped, palette='Greens')
# Titles and labels
plt.title('Profitability in 2020 by Country', fontsize=10)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Profitability (%)', fontsize=14)
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()
```

```python
# Convert column names to strings (just in case)
df_dropped.columns = df_dropped.columns.astype(str)
# Convert '2016' and '2018' to numeric (if needed)
df_dropped[['2019', '2022']] = df_dropped[['2019', '2022']].apply(pd.to_numeric, errors='coerce')
# Melt the DataFrame to long format
df_melted = df_dropped.melt(id_vars=['Country'], value_vars=['2019', '2022'],
                             var_name='Year', value_name='Profitability')
# Plot the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Profitability', hue='Year', data=df_melted, palette='viridis')
# Titles and labels
plt.title('Profitability in 2019 vs. 2022 by Country', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Profitability (%)', fontsize=14)
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()
```

```python
# Example: Plotting the HPI for a specific country over the years
country = 'Bulgaria'  # Replace with the country you're interested in
years = df_dropped.columns[1:]  # Assuming first column is 'Country'
values = df_dropped.loc[df_dropped['Country'] == country].values[0][1:]  # Grab the HPI values

# Plot
plt.figure(figsize=(10, 5))
plt.plot(years, values, marker='o', label=country)
plt.title(f'House Price Index for {country} (2010-2023)')
plt.xlabel('Year')
plt.ylabel('HPI')
plt.xticks(rotation=45)  # Rotate x-ticks if necessary
plt.axhline(100, color='grey', linewidth=0.8, linestyle='--')  # Reference line at HPI = 100
plt.legend()
plt.grid()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Create a DataFrame
df = pd.DataFrame(df_dropped)

# Convert DataFrame to long format for Seaborn
df_long = df.melt(id_vars=['Country'], var_name='Year', value_name='HPI')

# Convert 'Year' to integer for proper handling
df_long['Year'] = df_long['Year'].astype(int)

# Set the style for Seaborn
sns.set(style="whitegrid")

# Create a line plot for HPI over the years
plt.figure(figsize=(14, 8))
sns.lineplot(data=df_long, x='Year', y='HPI', hue='Country', marker='o')

# Adding titles and labels
plt.title('House Price Index (HPI) from 2010 to 2023 for Various Countries', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('House Price Index (HPI)', fontsize=14)
plt.axhline(100, color='grey', linewidth=0.8, linestyle='--')  # Optional reference line
plt.xticks(rotation=45)  # Rotate x labels if necessary
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
df = pd.DataFrame(df_dropped)

# Calculate Year-on-Year Percentage Change
hpi_changes = df.set_index('Country').pct_change(axis=1) * 100  # Percentage change

# Find the maximum increase for each country
max_increases = hpi_changes.max(axis=1)

# Create a DataFrame for easier sorting
max_increases_df = pd.DataFrame(max_increases, columns=['Max Increase (%)'])

# Sort and get top 10 countries with the highest increase
top_countries = max_increases_df.sort_values(by='Max Increase (%)', ascending=False).head(10)

top_countries
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=top_countries, x='Country', y='Max Increase (%)', palette='viridis')

# Adding titles and labels
plt.title('Top 10 Countries with the Highest Increase in House Price Index', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Max Increase (%)', fontsize=14)
plt.xticks(rotation=45)  # Rotate x-ticks for better visibility

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import requests

# Define the URL of the image
image_url = "https://ec.europa.eu/eurostat/documents/4187653/18051237/house-prices-rents-change-between-2010-q2-2024jpg.jpg/14395314-47a4-721e-37ef-ea84c8a10a28?t=1727940974598"

# Send a GET request to the image URL
response = requests.get(image_url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Open a file in binary write mode and save the image
    with open("house_prices_rents_change.jpg", "wb") as file:
        file.write(response.content)  # Write the image data to the file
    print("Image downloaded successfully.")
else:
    print(f"Failed to download image. Status code: {response.status_code}")
```
